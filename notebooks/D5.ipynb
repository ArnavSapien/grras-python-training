```python
def hi(a,b):
    print(a+b)
hi(5,19)
```

```python
def add(a=0,b=0): #default parameter
    print(a+b)
add()
```

```python
def power(a,b): 
    return a**b
print(power(b=2,a=3)) #Specific argument
```

```python
def student(*a): #Multiple parameters
    print(a)
student(1,2,3,4,5,6)
```

```python
def marks(a):
    print(a)
marks([10,20,30,40,50])
```

```python
def address(p):
    for i in p:
        for j in range(0,len(i)):
            print(i[j])
address(["Hello","Arnav"])
```

```python
def cal(a,b):
    return a+b,a-b
sum,dif=cal(5,10)
print(sum,dif)
```

```python
sum=lambda x,y: x+y
print(sum(1,2))
```
