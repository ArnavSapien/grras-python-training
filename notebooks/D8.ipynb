```python
#OOPs

class avs:
    def __init__(self,name): #constructor
        self.name=name
    def show(self):
        print(self.name)

a=avs("HI")
a.show()
```

```python
class other:
    def __init__(self):
        print("Constructor")
    def show(self,ob): # for a method it must consist (self,parameters........)
        print("Show",ob)
ot=other()
ot.show("HI")
```

```python
class dog:
    @staticmethod
    def show(): 
        print("Show")
d=dog()
d.show()
```

```python
def greet():
    return "Hello!"

def farewell():
    return "Goodbye!"

# Store functions in a dictionary
actions = {
    "hi": greet(),
    "bye": farewell()
}
print(actions["hi"])
```

```python
class student:
    def __init__(self,*args):
        print(type(args))
        self.name=args[0]
    def getStu(self):
        print("The Student name is :",self.name)
s=student("AVS","8080")
s.getStu()
```

```python
class emp:
    def __init__(self,name,deptId,salary):
        self.name=name
        self.deptId=deptId
        self.salary=salary
    def empD(self):
        print(self.name,self.deptId,self.salary)
    raiseSal= lambda self,x: setattr(self, "salary", self.salary * (x+100)/100)

e=emp("E1",101,85000)
e.raiseSal(10)
e.empD()
```

```python
class student:
    def __init__(self,*args):
        self.data=args
    def users(self):
        for i in self.data[0]:
            print(i)
    def details(self):
        for i in self.data[1]:
            print(i,"=",self.data[1][i])

s=student(["dheeraj","kunal","harsh","praveen"],{"address":"kukas","college":"arya","location":"Jaipur"})
s.users()
s.details()
```

```python
class new:
    def __init__(self,**args):
        self.data=args
    def profile(self,**kwargs):
        for key, value in kwargs.items():
            print(f"{key}: {value}")
n=new()
n.profile(name="AVS",age=20)
```
