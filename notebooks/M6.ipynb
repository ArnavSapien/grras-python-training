# Evaluation Metrics

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Actual values prediction
y_true = np.array([50, 60, 75, 90, 100])

# predicted value
y_pred = np.array([52, 58, 78, 88, 105])

# MAE (mean absolute error)
mae = mean_absolute_error(y_true, y_pred)
print("MAE: ", mae)

# MSE (mean squared error)
mse = mean_squared_error(y_true, y_pred)
print("MSE: ", mse)

# RMSE (root mean square error)
rmse = np.sqrt(mse)
print("RMSE: ", rmse)

# R2 score
r2s = r2_score(y_true, y_pred)
print("R2s: ", r2s)
```

# Lasso vs Ridge

## Ridge

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn. linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error

data = {
    "size": [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2700, 3000, 3200],
    "bedrooms": [2, 2, 3, 3, 3, 4, 4, 4, 5, 5],
    "age": [20, 18, 15, 12, 10, 8, 6, 5, 4, 2],
    "price": [30, 35, 45, 50, 55, 60, 68, 72, 80, 85]
}

df = pd.DataFrame(data)
print(df)

# split data
x = df[['size','bedrooms','age']]
y = df['price']

# split data in train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# training (ridge)
ridge = Ridge(alpha=1.0)
ridge.fit(x_train,y_train)

# prediction
ridge_p = ridge.predict(x_test)
print("y_test\n",y_test)
print("prediction\n",ridge_p)

# errors
mae = mean_absolute_error(y_test,ridge_p)
mse = mean_squared_error(y_test,ridge_p)
print(f"MAE: {mae}\nMSE: {mse}")

# model intercept
print(ridge.intercept_)
 
# model coef  price = intercept + (cof1,* feat1)
print(ridge.coef_)
```

## Lasso

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn. linear_model import Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error

data = {
    "size": [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2700, 3000, 3200],
    "bedrooms": [2, 2, 3, 3, 3, 4, 4, 4, 5, 5],
    "age": [20, 18, 15, 12, 10, 8, 6, 5, 4, 2],
    "price": [30, 35, 45, 50, 55, 60, 68, 72, 80, 85]
}

df = pd.DataFrame(data)
print(df)

# split data
x = df[['size', 'bedrooms', 'age']]
y = df['price']

# split data in train and test
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)

# training (ridge)
lasso = Lasso(alpha=1.0)
lasso.fit(x_train, y_train)

# prediction
lasso_p = lasso.predict(x_test)
print("y_test\n", y_test)
print("prediction\n", lasso_p)

# errors
mae = mean_absolute_error(y_test, lasso_p)
mse = mean_squared_error(y_test, lasso_p)
print(f"MAE: {mae}\nMSE: {mse}")

# model intercept
print(lasso.intercept_)

# model coef  price = intercept + (cof1 * feat1) + (cof2 * feat2) +...
print(lasso.coef_)
```

# Logistic Regression

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression as LR

data = {
    "hours": [1, 2, 3, 4, 5, 6, 7],
    "result": [0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# split x and y
x = df[['hours']]
y = df['result']

# model
model = LR()
model.fit(x, y)

# prediction
new_data = pd.DataFrame({
    "hours": [3]
})

pred_d = model.predict(new_data)
print(pred_d)

# probability prediction
prob_d = model.predict_proba(new_data)
for i in prob_d[0]:
    print(np.round(i*100,2),"%")
```

# KNN (K-Nearest Neighbors)
