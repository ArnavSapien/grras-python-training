**Sort**

```python
#1d
import numpy as np
arr=np.array([10,5,20,30,9,8,7])
print(arr)
print("Sorted ascending:",np.sort(arr))
print("Sorted descending:",np.sort(arr)[::-1])

arr2=np.array([[2,9],[10,1]])
print("Sorted row wise:\n",np.sort(arr2,axis=1)) #by default sort is row wise i.e. axis 1
print("Sorted column wise:\n",(np.sort(arr2,axis=0)))
print("Sorting ascending:\n",np.sort(np.sort(arr2,axis=0)))
print("Sorting descending:\n", np.sort(np.sort(arr2, axis=0))[::-1, ::-1])
```

```python
#filter
arr=np.array([10,20,30,40])
print(arr)
print(arr[arr<=20])
```

```python
#Fancy indexing / BOOLEAN INDEXING
import numpy as np
arr = np.array([10,20,3,4,5,90,88,30])
print(arr)
arr_Fancy = arr[[0,2]]
print(arr_Fancy)
```

```python
#np.where
import numpy as np

arr = np.array([10,3,4,9,5,20,30,70,55])
print(arr)
arrW = np.where(arr>40,arr,"")
print(arrW)
arrW = np.where(arr>40,arr+2,arr-2)
print(arrW)
```

```python
#Concatenation: adding is not same as concatenation
import numpy as np
arr=np.array([10,20,30])
brr=np.array([1,2,3])
print("Concatenated:\n", np.concatenate((arr, brr)))

a=np.array([[10,20],[30,40]])
b=np.array([[1,2],[3,4]])
print("Concatenated:\n", np.concatenate((a, b),axis=1))
```

```python
import numpy as np

arr = np.array([10,3,4,9,5,20,30,70,55])
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.min(arr),np.argmin(arr))
print(np.max(arr),np.argmax(arr))
print(np.var(arr))
print(np.std(arr))

"""
1. np.sum(a) -> Sum of all elements
2. np.mean(a) -> Average = Sum of elements / Number of elements
3. np.median(a) -> Middle value after sorting
4. np.min(a) -> Smallest value in array
5. np.max(a) -> Largest value in array
6. np.var(a) mean,difference,square,average | (sum = ddof) -> divide by array length
7. np.std(a) -> Standard Deviation = √Variance -> Measures spread of data
8. np.prod(a) -> Multiplication of all elements
9. np.cumsum(a) -> Cumulative (running) sum
10. np.cumprod(a) -> Cumulative (running) product
11. np.argmin(a) -> Index position of minimum value
12. np.argmax(a) -> Index position of maximum value
13. np.abs(a) -> Converts negative values to positive -> Absolute value (distance from zero)
14. np.unique(a) -> Returns unique values only -> Removes duplicates
15. np.percentile(a, 50) -> Finds percentage-based value -> 50th percentile = Median
16. np.quantile(a, 0.5) -> Similar to percentile -> 0.5 = 50%
17. np.ptp(a) -> Range = Max - Min
18. np.any(a) -> True if at least one value is True

"""
```
