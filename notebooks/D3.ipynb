```python
dict={
    "movies":["Iron Man","Pirates Of Carribean","Joker"]
}

for i in [0,1,2]:
    dict["movies"][i]=input(f"Enter movie name{i} :")

print("Movies:")
for movies in dict["movies"]:
    print(movies)

s=input("Enter movie to search :")

if s in dict["movies"]:
    print("Movie present")
else:
    print("Movie not found")
```

```python
dict={
    "cars":{
        "Mahindra":"XUV",
        "Honda":"city"
    }
}
for keys in dict["cars"]:
    dict["cars"][keys]=input(f"Enter model :")
print(dict)
s=input("Enter car to search :")
for keys in dict["cars"]:
    if s in dict["cars"][keys]:
        print("Car present")
```
