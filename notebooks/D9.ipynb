```python
class Address:
    def __init__(self,city,state):
        self.city=city
        self.state=state
    def show(self):
        print("The city is",self.city)
a=Address("Jaipur","Rajasthan")
a.show()
```

```python
#Inheritance
class Address:
    def __init__(self,city,state):
        self.city=city
        self.state=state
    def location(self):
        return f"My location is {self.city} in {self.state}"

class User(Address):
    def __init__(self,name,age,city,state):
        super().__init__(city,state)
        self.name=name
        self.age=age
        #self.city=city
        #self.state=state
    def userName(self):
        print("Your name is",self.state)
    def userDetails(self):
        print(self.name,self.city,self.state,self.age)

a=User("Arnav",20,"Kukas","Rajasthan")
a.userDetails()
a.location()
```

```python
#Encapsulation
class Address:
    def __init__(self,city,state):
        self.__city=city #private attribute cannot be directly accessed
        self.state=state
    def location(self):
        print(f"my location is {self.__city} in {self.state}")
a=Address("Jaipur","Rajasthan")
a.location()
```

```python
#Polymorphosm
class Address:
    def __init__(self,city,state):
        self.city=city
        self.state=state
    def location(self):
        print(f"My address location is {self.city} in {self.state}")
class homeTown:
    def __init__(self,city,state):
        self.city=city
        self.state=state
    def location(self):
        print(f"My hometown location is {self.city} in {self.state}")

a=Address("vadodara","gujarat")
a.location()
b=homeTown("jaipur","rajasthan")
b.location()
```

```python
#overriding
class address:
    def __init__(self,city):
        self.city=city
    def location(self):
        print(self.city)
class user(address):
    def __init__(self,city):
        super().__init__(city)
    def location(self):
        print("user location is :",self.city)
a=user("jaipur")
a.location()
```
