```python
#map
def square(x): return x*x
l=[10,20,30]
l1=list(map(square,l))
print(l1)
```

```python
m=["hi",'hello']
n=list(map(str.upper,m))
print(n)
```

```python
l=[10,20,30]
l1=list(map(lambda x:x*x,l))
print(l1)
```

```python
#filter
def helo(x):
    return x.startswith("a")
l=["apple","banana","cat","dog"]
result = list(filter(helo, l))
print(result)
```

```python
l1=[]
for i in l:
    if 'a' == i[0]:
        l1.append(i)
print(l1)
```

```python
#try except
try:
    num1=10
    num2=0
    print(num1/num2)
except:
    print("error")
```

```python
#try except final
try:
    num1=10
    num2=0
    print(num1/num2)
except:
    print("error")
finally:
    print("this is finally block")
```
