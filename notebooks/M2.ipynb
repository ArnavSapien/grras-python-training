Handling Missing Values

```python
import pandas as pd
import numpy as np
data = {
    "colors":['red','green','blue','orange','green','blue',np.nan]
}
df = pd.DataFrame(data)
print(df)
```

Handle null values

```python
df.dropna(inplace=True)
print(df)
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder as le
df['color_encoder'] = le().fit_transform(df['colors'])
print(df)

# or
# from sklearn.preprocessing import LabelEncoder
# le=LabelEncoder()
# df['color_encoder'] = le.fit_transform(df['colors'])
# print(df)

# or
# import sklearn
# df['color_encoder'] = sklearn.preprocessing.LabelEncoder().fit_transform(df['colors'])
# print(df)
```

One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder

# one hot encoding (by the help of sklearn)
ohe = OneHotEncoder()
encoded = ohe.fit_transform(df[['colors' ]])
print (encoded.toarray(),'\n\n')

encoder_df =pd.get_dummies(df.colors)
print(encoder_df)
```

```python
# combination of handle missing value | label encoding | onehot encoding
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder , OneHotEncoder

data = {
"age": [10,20,np.nan, 26, 30],
"gender": ['male', 'female', 'others', 'male',np.nan],
"name": ['dheeraj ', 'kavi', 'sapu', 'yash', 'hello']
}
df = pd.DataFrame(data)

# handle missing values
df['age' ] = df['age'].fillna(df['age'].mean())
df.dropna(subset=['gender'], inplace=True)
print(df,'\n\n')

# label encoding
le = LabelEncoder()
df['gender_label_encoded'] = le.fit_transform(df['gender'])
print(df,'\n\n')

#one hot encoding
oe = OneHotEncoder()
oe_e = oe.fit_transform(df[['gender']]).toarray()
print('One Hot Encoder\n',oe_e,'\n\n')
```
