Working with Date Values

```python
import pandas as pd

url = "../datasets/file2.json"
df = pd.read_json(url)

df["doj"] = [
    "2025-01-10",
    "2025-02-10",
    "2025-03-10",
    "2025-04-10",
    "2025-05-10"
]

df["doj"].dtype
```

```python
df["doj"] = pd.to_datetime(df["doj"])

df["year"] = df["doj"].dt.year
df["month"] = df["doj"].dt.month
df["day"] = df["doj"].dt.day
df["day_name"] = df["doj"].dt.day_name()

df
```

```python
df["date_after_20_days"] = df["doj"] + pd.to_timedelta(20, unit="D")

df
```
