# Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
 
data = {
    "age":[22,25,24,45,50,48,20,47],
    "income":[22000,25000,24000,45000,50000,48000,20000,47000]
}
 
df = pd.DataFrame(data)
print(df)
 
# input | feature
x = df[['age','income']]
 
# model
model = KMeans(n_clusters=2,random_state=42) # n_clusters -> 2
model.fit(x)
 
# clusters
clusters = model.labels_ # 0 is defining the young age and 1 is define the older age
print(clusters)
df['groups'] = clusters # add new cols
print(df)
```

## example

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
 
data = {
    "age":[22,25,24,45,50,48,20,47],
    "income":[22000,25000,24000,45000,50000,48000,20000,47000]
}
 
df = pd.DataFrame(data)
print(df)
 
x = df[['age','income']]
 
model = KMeans(n_clusters=2,random_state=42)
model.fit(x)
 
df['cluster'] = model.labels_
 
plt.scatter(df['age'],df['income'],c=df['cluster'])
plt.xlabel('age')
plt.ylabel('income')
plt.title('Group the Data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = {
    "sleep_hours": [5, 6, 7, 8, 9, 4, 3, 2, 10, 11],
    "productivity": [70, 80, 85, 95, 100,35, 30, 20, 100, 100]
}
df = pd.DataFrame(data)

x = df[['sleep_hours', 'productivity']]
model = KMeans(n_clusters=2, random_state=42)
model.fit(x)

df['cluster'] = model.labels_

plt.scatter(df['sleep_hours'], df['productivity'], c=df['cluster'])
plt.xlabel('sleep_hours')
plt.ylabel('productivity')
plt.title('Sleep Output')
plt.show()
```
